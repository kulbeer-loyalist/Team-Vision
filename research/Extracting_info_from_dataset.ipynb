```python
import pandas as pd
import cv2
from pathlib import Path

video_folder = r"C:\Users\edulo\OneDrive\Documentos\Loyalist\Term 2\Step Presentation - Project\SL Project\WLASL_videos\WLASL_videos - Original"

# Creating an empty dataframe to store the videos informations
video_info_df = pd.DataFrame(columns=['Name', 'Date', 'Length', 'Size', 'Type', 'FrameRate', 'Bitrate', 'Frames', 'Resolution'])

# Iterating over each video file in the folder
# Reference: https://www.futurelearn.com/info/courses/introduction-to-image-analysis-for-plant-phenotyping/0/steps/305359

for video_path in Path(video_folder).glob('*.mp4'):
    try:

# Opening the video using OpenCV and getting the video informations
        video = cv2.VideoCapture(str(video_path))
        name = video_path.stem
        date = video_path.stat().st_mtime
        length = video.get(cv2.CAP_PROP_FRAME_COUNT) / video.get(cv2.CAP_PROP_FPS)
        size = video_path.stat().st_size
        video_type = 'MP4'
        framerate = video.get(cv2.CAP_PROP_FPS)
        bitrate = video.get(cv2.CAP_PROP_BITRATE)
        frames = video.get(cv2.CAP_PROP_FRAME_COUNT)
        resolution = (int(video.get(cv2.CAP_PROP_FRAME_WIDTH)), int(video.get(cv2.CAP_PROP_FRAME_HEIGHT)))

# Append video information to the dataframe
        video_info_df = video_info_df.append({'Name': name,
                                              'Date': date,
                                              'Length': length,
                                              'Size': size,
                                              'Type': video_type,
                                              'FrameRate': framerate,
                                              'Bitrate': bitrate,
                                              'Frames': frames,
                                              'Resolution': resolution},
                                             ignore_index=True)

        video.release()
    except Exception as e:
        print(f"Error video file: {video_path} ({e})")

# Converting the date column to datetime format
# Reference: https://pandas.pydata.org/docs/reference/api/pandas.to_datetime.html
video_info_df['Date'] = pd.to_datetime(video_info_df['Date'], unit='s')

# Display the dataframe
video_info_df
```

```python
video_info_df.to_csv("video_info_df.csv")
```
